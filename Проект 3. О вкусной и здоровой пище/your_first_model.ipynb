Преобразуем Price Range и пустые значения заменяем средним

```python
def range_changer(row):
    if row['Price Range'] == '$':
        return 1
    elif  row['Price Range'] == '$$ - $$$':
        return 2
    elif  row['Price Range'] == '$$$$':
        return 3
    else:
        return None
    
df['price_range'] = df.apply(lambda row: range_changer(row), axis=1)

df['price_range'] = df['price_range'].fillna(np.mean(df['price_range']))
df[['price_range']].info()
```

Заполняем пустые Number of Reviews

```python
df['Number of Reviews'] = df['Number of Reviews'].fillna(np.median(df['Number of Reviews'].dropna()))
```

Создаём отдельный столбик только для отзывов и проверяем на положительные и отрицательные отзывы

```python
df.Reviews = df.Reviews.str.strip('[]').str.replace('[','').str.replace(']','').str.split(',')
```

```python
# Отдельный столбик только для текста отзывов
reviews = []
for list1 in df['Reviews']:
    reviews.append(list1[0:2])

df['reviews'] = pd.Series(reviews)
```

```python
df['reviews']
```

```python
# Bсе отзывы по рестарану в одну строчку и разделяем на отдельные слова
row_all = list()

for row in df['reviews']:
    row_t = []
    row_t = row[0].lower().split(' ') + row[1].lower().split(' ')
    row_all.append(row_t)
    row_t
    
df['reviews_list'] = pd.Series(row_all)
```

```python
# Проверяем в каких коментариях есть положительные или отрицательные слова

good_words = ['good', 'excellent', 'tasty', 'best', 'nice', 'beautiful', 'unique', 'great', 'lovely', 'brilliant']
bad_words = ['not', 'bad', 'seriously',"'don't'", 'horrible', 'ok', 'terrible', 'but', 'disappointed', 'enough']
is_good = []
is_bad = []

for review in df['reviews_list']:
    for word in review:
        if word in good_words:
            is_good.append(1)
        else:
            is_good.append(0)
        if word in bad_words:
            is_bad.append(1)
        else:
            is_bad.append(0)
                
df['is_good'] = pd.Series(is_good)      
df['is_bad'] = pd.Series(is_bad)
```

Создаём отдельный столбец для даты отзывов и вычисляем как давно был написан отзыв

```python
dates = []
for list1 in df['Reviews']:
    if len(list1) == 4:
        date = list1[-2:]
    elif len(list1) in [2, 3]:
        date = list1[-1]
    else:
        date = list1[-2:]
    dates.append(date)

df['dates'] = pd.Series(dates)
#print(dt.datetime.strptime(df['dates'][0][0][2:-1], '%m/%d/%Y'))
df['dates']
```

```python
# находим дату из последнего столбика
last_date = []

for row in df['dates']:
    dates_diff = []
    if len(row) == 2:
        last_date.append(dt.datetime.strptime(row[-1][2:-1], '%m/%d/%Y'))
    else:
        last_date.append(None)

last_date
```

```python
# Смотрим как давно был оставлен отзыв
today = dt.datetime.today()
day_diff = []
for date in last_date:
    if date != None:
        diff = (today-date).days
        day_diff.append(diff)
    else:
        day_diff.append(None)
        
df['days_since_review'] = pd.Series(day_diff)
df['days_since_review'] = round(df['days_since_review'].fillna(np.mean(df['days_since_review'].dropna())))
df['days_since_review']
```

```python
# Отдельный столбик только для даты отзывов
dates = []
for list1 in df['Reviews']:
    dates.append(list1[2:4])

df['dates'] = pd.Series(dates)
dt.datetime.strptime(df['dates'][0][0][2:-1], '%m/%d/%Y')
```

Обьединяем dummies и основной df в одно целое. Заполняем пустые dummies нулями

```python
df = pd.concat([df, dummies_city], axis=1)
df = df.fillna(0)
df.head()
```

```python
df_num_only = df.drop(['Restaurant_id', 'Price Range', 'Cuisine Style', 'City', 'URL_TA', 'ID_TA', 'Reviews', 'reviews', 'reviews_list', 'dates'], axis=1)
df_num_only.info()
```

# Разбиваем датафрейм на части, необходимые для обучения и тестирования модели

```python
# testing my vers
X = df_num_only.drop(['Rating'], axis = 1)
y = df_num_only['Rating']
```

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
############## X = df.drop(['Restaurant_id', 'Rating'], axis = 1)
############### y = df['Rating']
```

```python
# Загружаем специальный инструмент для разбивки:
from sklearn.model_selection import train_test_split
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Создаём, обучаем и тестируем модель

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.

print('MAE:', round(metrics.mean_absolute_error(y_test, y_pred), 5))
list_of_tries.append(round(metrics.mean_absolute_error(y_test, y_pred), 5))
print(list_of_tries)
```
