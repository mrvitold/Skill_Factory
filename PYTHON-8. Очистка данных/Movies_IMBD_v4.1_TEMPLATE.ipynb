# 21. В каком месяце за все годы суммарно вышло больше всего фильмов?

```python
answers['21'] = 'Сентябрь'
```

```python
date1 = data['release_date'].apply( lambda x: x.split('/'))  # разделяем  дату на год, месяц и день
months = []  # Создаём список месяцев выхода фильмов при помощи for цикла
for i in date1:
    months.append(i[0])

month = Counter(months).keys()  # записявем уникальные месяцы
count = Counter(months).values()  # считаем сколько раз упоминаетя каждый месяц
month_count = pd.DataFrame(zip(month, count), columns = ['month', 'count'])  # Создаём датафрейм с месяцами и их количеством в датафрейме
month_count.sort_values(by='count', ascending=False)
```

# 22. Сколько суммарно вышло фильмов летом? (за июнь, июль, август)

```python
answers['22'] = 450
```

```python
date1 = data['release_date'].apply( lambda x: x.split('/'))
months = []  # # Создаём список месяцев выхода фильмов при помощи for цикла, в который попадают только летние месяцы
for i in date1:
    if int(i[0]) in [6, 7, 8]:
        months.append(i[0])

month = Counter(months).keys()
count = Counter(months).values()
month_count = pd.DataFrame(zip(month, count), columns = ['month', 'count'])
month_count['count'].sum()
```

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


ВАРИАНТ 2

```python
answers['27'] = 'Daniel Radcliffe & Rupert Grint'
```

```python
import itertools

data_actors = data.copy()
data_actors['cast'] = data_actors['cast'].str.split('|') 
data_actors.cast = data_actors.cast.apply(lambda x: list(itertools.combinations(x, 2)))  # создаём пары в ячейке актёров.
data_pairs = data_actors.explode('cast')  # Длы каждой пары делаем отдельную строчку
data_pairs.cast.value_counts(ascending=False)
```
